```python
import os
# import shutil
import numpy as np
import cProfile
from PIL import Image
```

```python
#obtenir l'ensemble des labels de la base de données 

def get_classes(file_name, simplified=True):

    classes = {}
    file = open(file_name, 'r')
    i = 0
    file.readline()
    file.readline()
    while True:

        line = file.readline()

        if not line:
            file.close()
            break

        if simplified :
            
            classes[line.split(" ")[0]] = line.split(" ")[-1][0]
            
        else : 
            
            classes[line.split(" ")[0]] = i
        i = i + 1

    return classes


```

```python
#modifier les path en fonction 

bbox_train_file = "../../datasets/Category and Attribute Prediction Benchmark/train_bbox.txt"
train_file = "../../datasets/Category and Attribute Prediction Benchmark/train.txt"

bbox_val_file = "../../datasets/Category and Attribute Prediction Benchmark/val_bbox.txt"
val_file = "../../datasets/Category and Attribute Prediction Benchmark/val.txt"


bbox_test_file = "../../datasets/Category and Attribute Prediction Benchmark/test_bbox.txt"
test_file = "../../datasets/Category and Attribute Prediction Benchmark/test.txt"


categories_file =  "../../datasets/Category and Attribute Prediction Benchmark/list_category_cloth.txt"
```

```python
#créait une structure de données utilisables par le modèle yolov5
#paramètre pour pouvoir changer que les labels 


def image_resizig(image,bbox, target_size = 300, fill_color=(0, 0, 0)):
    
    image_w, image_h = image.size 
    
    #create background
    new_image = Image.new('RGB', (target_size, target_size), fill_color)
    
    #add image, image cropped if dimensions > (300, 300)
    new_image.paste(image.crop((0, 0, min(target_size, image_w), min(target_size, image_h))),
                    (int((target_size - min(target_size, image_w)) / 2), int((target_size - min(target_size, image_h)) / 2)))                          
    
    
    #calculus parameters mandatory for IA 
    b_center_x = (int(bbox[0]) + int(bbox[2])) / 2 + int((target_size - min(target_size, image_w)) / 2)
    b_center_y = (int(bbox[1]) + int(bbox[3])) / 2 + int((target_size - min(target_size, image_h)) / 2)
    b_width    = (int(bbox[2]) - int(bbox[0])) / 2 
    b_height   = (int(bbox[3]) - int(bbox[1])) / 2 


    b_center_x /= target_size
    b_center_y /= target_size
    b_width    /= target_size
    b_height   /= target_size
    

     
    
    return new_image, b_center_x, b_center_y, b_width, b_height




def dataset_creation(bbox_file_name, location_file_name,categories, parameter = "train", limite = np.inf): 

 
    bbox_file = open(bbox_file_name, 'r')
    location_file = open(location_file_name, 'r')
    
    bbox_lines = bbox_file.readlines()
    files_information = location_file.readlines()

    if not(len(bbox_lines) == len(files_information)):
        print("erreur !")
        bbox_file.close()
        location_file.close()
        return
        
        

    for index in range(len(bbox_lines)):

        


        #categorie du vêtement
        clothe_type = files_information[index].split("/")[1].split("_")[-1]
        #où trouver le fichier image
        image_file_name = files_information[index][:-1]
        #nom pour sauvergarder le fichier labels et l'image dans train
        target_file_name = files_information[index][:-1].split("/")[-2] + "_" + files_information[index][:-1].split("/")[-1][:-4]

        image = Image.open("../../datasets/" + image_file_name)

        #bbox 
        bbox = bbox_lines[index].split(" ")[:-1]

       
        image_resized, b_center_x, b_center_y, b_width, b_height = image_resizig(image,bbox)

        image.close(); 

        #label pour yolo
        label = []
        label.append("{} {:.3f} {:.3f} {:.3f} {:.3f}".format(categories[clothe_type], b_center_x, b_center_y, b_width, b_height))


        
        with open( "../../datasets/FashionIA/" + parameter + "/labels/" + target_file_name + ".txt", 'w') as f:
            f.write('\n'.join(label))


#         shutil.copy("../../datasets/" + image_file_name, "../../datasets/FashionIA/" + parameter + "/images/" + txt_file_name + ".jpg")
        image_resized = image_resized.save("../../datasets/FashionIA/" + parameter + "/images/" + target_file_name + ".jpg")
        
        
        index = index + 1
        
        if index > limite : 
            
            break 





    bbox_file.close()
    location_file.close()
    print(parameter + "  : finished")

```

```python
categories = get_classes(categories_file)
```

```python
dataset_creation(bbox_train_file,train_file,categories, parameter = 'train')
dataset_creation(bbox_val_file,val_file,categories, parameter = 'val')
dataset_creation(bbox_test_file,test_file,categories, parameter = 'test')
```
